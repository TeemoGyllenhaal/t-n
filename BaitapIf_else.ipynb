```python
import math
a = float(input("Nhập giá trị a: "))
b = float(input("Nhập giá trị b: "))
c = float(input("Nhập giá trị c: "))
if a == 0:
    if b == 0:
        if c==0:
            print ("Phương trình có vô số nghiệm")
        else:
            print ("Phương trình vô nghiệm ")
    else:
        print ("Phương trình có nghiệm x = ", -c/b) 
else:
    delta = b*b - 4 * a * c
    if delta < 0:
        print ("Phương trình vô nghiệm ")
    else:
        if delta == 0:
            print("Phương trình có nghiệm kép x1 = x2 = ", -b/2*a)   
        else:
            x1 = (-b + math.sqrt(delta))/2*a
            x2 = (-b - math.sqrt(delta))/2*a
            print("Phương trình có 2 nghiệm x1 = ",x1,",x2 = ",x2)
```

```python
from math import sqrt, cos, acos, pi 
a = float(input("Nhập giá trị a: "))
b = float(input("Nhập giá trị b: "))
c = float(input("Nhập giá trị c: "))
d = float(input("Nhập giá trị d: "))
if a == 0:
    if b == 0:
        if c == 0:
            if d == 0:
                print ("Phương trình có vô số nghiệm")
            else:
                print ("Phương trình vô nghiệm ")
        else:
            print ("Phương trình có nghiệm x = ", -d/c)
    else:
        delta = c*c - 4 * b * d
        if delta < 0:
            print ("Phương trình vô nghiệm ")
        else:
            if delta == 0:
                print("Phương trình có nghiệm kép x1 = x2 = ", -c/2*b)   
            else:
                x1 = (-c + sqrt(delta))/2*b
                x2 = (-c - sqrt(delta))/2*b
                print("Phương trình có 2 nghiệm x1 = ",x1,",x2 = ",x2)
else:   
    AltDelta = (b**2)-(3*a*c)
    if AltDelta > 0:
        k = (9*a*b*c-2*b**3-27*a**2*d)/(2*sqrt(abs(AltDelta)**3))
        if abs(k) <= 1:
            x1 = (2*sqrt(AltDelta)*cos(acos(k)/3)-b)/(3*a)
            x2 = (2*sqrt(AltDelta)*cos(acos(k)/3-2*pi/3)-b)/(3*a)
            x3 = (2*sqrt(AltDelta)*cos(acos(k)/3+2*pi/3)-b)/(3*a)
            print("Phương trình có 3 nghiệm:")
            print("x1 = ",x1)
            print("x2 = ",x2)
            print("x3 = ",x3)
        else:
            x = ((sqrt(AltDelta)*abs(k))/(3*a*k))*((abs(k)+sqrt((k**2)-1))**(1/3)+(abs(k)-sqrt((k**2)-1))**(1/3))-(b/(3*a))
            print("Phương trình có 1 nghiệm duy nhất: x = ",x)

    else:
        if AltDelta == 0:
            x = (-b+(b**3-27*a**2*d)**1/3)/3*a
            print("Phương trình có 1 nghiệm bội: x = ",x)
        else:
            k = (9*a*b*c-2*b**3-27*a**2*d)/(2*sqrt(abs(AltDelta)**3))
            x = (sqrt(abs(AltDelta))/(3*a))*((k+sqrt((k**2)+1))**(1/3)+-abs((k-sqrt((k**2)+1)))**(1/3))-(b/(3*a))
            print("Phương trình có 1 nghiệm duy nhất: x = ",x)
```
